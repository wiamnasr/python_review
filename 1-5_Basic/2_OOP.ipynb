### Attributes and the Class keyword:

```python
# Creating a simple class
class Sample():
    pass

# creating an instance of the sample class
my_sample = Sample()

type(my_sample)
```

```python
# circle class
class Circle():

    # Class Object Attribute
    pi = 3.14

    def __init__(self,radius=1):

        self.radius = radius

        self.area = radius*radius*self.pi
        # since its a class object attribute, the above can be referred to as Circle.pi instead of self.pi -> it will be the same regardless of the instance of the class

    # Method
    def get_circumference(self):
        return self.radius * self.pi * 2
```

```python
# creating an instance of the circle:
# over-writing the default value of the radius with 30 instead of 1
my_circle = Circle(30)
```

```python
my_circle.pi
```

```python
my_circle.radius
```

```python
my_circle.area
```

```python
my_circle.get_circumference()
```

### Inheritance

##### Inheritance is a way to form new classes using classes that have already been defined -> Code re-usability and complexity reduction

```python
# Creating a base class
class Animal():
    def __init__(self):
        print("Animal Created")

    def who_am_i(self):
        print("I am an animal")

    def eat(self):
        print("I am eating")


# Newly formed classes, can use the Animal class in order to inherit some of its methods that we may want to use again

# re-creating the Dog class, passing in Animal in order to inherit from the base class
# This is called a derived class
class Dog(Animal):

    def __init__(self):
        # Creating an instance of the Animal class when we are creating an instance of the Dog class
        # This is possible because we are inheriting from the Animal class
        Animal.__init__(self)
        print("Dog Created")

    # changing the inherited method for the Dog class:
    def who_am_i(self):
        print("I am a Dog!")

    def bark(self):
        print("Woof!")
```

```python
my_dog = Dog()
```

```python
# using the methods derived from the base class Animal:
my_dog.eat()
```

```python
# This method was initially inherited, but then we re-wrote it for the Dog class
my_dog.who_am_i()
```

```python
my_dog.bark()
```

### Polymorphism

##### Polymorphism refers to the way in which different object classes can share the same method name

##### Those methods can be called from the same place even though different objects might be passed in

```python
# Creating 2 new classes, redefining the Dog class and creating a new Cat class:

class Dog():
    def __init__(self, name):
        self.name =name

    def speak(self):
        return self.name + " says Woof! "

class Cat():
    def __init__(self, name):
        self.name =name

    def speak(self):
        return self.name + " says Meow! "


# creating 2 instances:
niko =Dog("niko")
felix = Cat("felix")

felix.speak()
```

```python
niko.speak()
```

```python
# demonstrating polymorphism with a for loop
# Both niko and felix share the same method name, called speak, however, in the outcome there are different types (__main__.Dog and __main__.Cat)
for pet in [niko, felix]:
    print(type(pet))
    print(type(pet.speak()))
```

```python
# demonstrating polymorphism using a function:
def pet_speak(pet):
    print(pet.speak())

pet_speak(niko)
```

```python
# demonstrating polymorphism using abstract classes and inheritance
# Abstract classes do not expect to be instanciated -> Designed to only serve as a base class
class Animal():
    def __init__(self,name):
        self.name = name

    def speak(self):
        # we will raise an error
        raise NotImplementedError("Subclass must implement this abstract method")

# If we try to create an instance of Animal, something not expected to be done, this will return the raised error, because its expecting us to inherit the Animal class then over-write the speak method

# Instead we call Dog and pass in Animal as we construct it
class Dog(Animal):
    def speak(self):
        return self.name + " says woof!"

class Cat(Animal):
    def speak(self):
        return self.name + " says meow!"


fido = Dog("Fido")
mojo = Cat("Mojo")

print(fido.speak())
print(mojo.speak())
```

### Special (Magic/Dunder) Methods with our own created objects

#### They use double underscores (__init__ was the first special method, called automatically when we create an object)

```python
class Book():

    def __init__(self, title, author, pages):
        self.title = title
        self.author =author
        self.pages = pages

    # __str__ special method: if any function asks for a string representation for the book class
    def __str__(self):
        # we return what we want to be printed here
        return f"{self.title} by {self.author}"

    # same can be done for length with the __len__
    def __len__(self):
        return self.pages

    # deleting a book
    def __del__(self):
        print("A Book Object has been deleted")


# creating an instance of the Book
b = Book("Machine Learning", "a famous author", 200)

# This will ask for a string representation of the instantiated Book
print(b)

len(b)
```

```python
# for deleting a book
del b
```
